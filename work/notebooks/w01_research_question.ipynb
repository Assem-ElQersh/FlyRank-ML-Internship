## 3. Quick look at the data (2-3 real numbers)

Let's look at the sheer volume of pages that are losing traffic. This proves why a manual review is impossible and why we need an automated scoring system.

```python
import pandas as pd
df = pd.read_csv('../../data/raw/content_refresh_anonymized.csv')

declining = len(df[df['trend_direction'] == 'down'])
print(f"Declining pages: {declining:,} out of {len(df):,} ({(declining/len(df))*100:.1f}%)")

meaningful_decline = len(df[(df['trend_direction'] == 'down') & (df['impressions_90d'] > 500)])
print(f"Declining pages with >500 impressions: {meaningful_decline:,}")

print("\nThis proves the scale of the problem: human editors cannot manually review thousands of declining pages. We need a ranked queue.")
```
